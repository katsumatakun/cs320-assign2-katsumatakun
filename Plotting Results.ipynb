```python
import matplotlib.pyplot as plt
import pandas as pd
def t100(n): 
    return 100*n 
  

frames = [i for i in range(4,129,4)]

f1 = open("FIFO-.txt", "r")
f1_line = f1.readline().split(" ")
f1_line.pop()
f1_line = list(map(float, f1_line))
f1_line = list(map(t100, f1_line)) 

lru1 = open("LRU-.txt", "r")
lru1_line = lru1.readline().split(" ")
lru1_line.pop()
lru1_line = list(map(float, lru1_line))
lru1_line = list(map(t100,lru1_line))

r1 = open("R-.txt", "r")
r1_line = r1.readline().split(" ")
r1_line.pop()
r1_line = list(map(float, r1_line)) 
r1_line = list(map(t100, r1_line))

sc1 = open("SC-.txt", "r")
sc1_line = sc1.readline().split(" ")
sc1_line.pop()
sc1_line = list(map(float, sc1_line)) 
sc1_line = list(map(t100, sc1_line))

opt1 = open("OPT-.txt", "r")
opt1_line = opt1.readline().split(" ")
opt1_line.pop()
opt1_line = list(map(float, opt1_line)) 
opt1_line = list(map(t100, opt1_line))

df_noloc = pd.DataFrame(index=frames)
df_noloc["OPT"] = opt1_line
df_noloc["SC"] = sc1_line
df_noloc["R"] = r1_line
df_noloc["LRU"] = lru1_line
df_noloc["FIFO"] = f1_line


fig = plt.figure()
ax1 = fig.add_subplot(111)
ax1.scatter(frames,f1_line , s=10,  marker="s", label='FIFO')
ax1.scatter(frames,r1_line, s=10,  marker="o", label='Random')
ax1.scatter(frames,lru1_line, s=10,  marker="^", label='LRU')
ax1.scatter(frames,sc1_line, s=10,  marker="<", label='SC')
ax1.scatter(frames,opt1_line, s=10,  marker="p", label='Optimal')
plt.legend(loc='upper left');
plt.ylabel("Hit Rate(%)")
plt.xlabel("Number of Frames")
# plt.show()

df_noloc
```
